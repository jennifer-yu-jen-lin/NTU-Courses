### Box Plot

```python
M = sns.boxplot(data=museum)
# abbreviate the name of the museums fot tick labels
tick_labels=['1 ATIC','2 AA','3 CAM','4 GtNC','5 FM','6 HQ','7 I','8 PH','9 ETG','10 MoSJ','11 BG']
M.set_xticklabels(tick_labels,rotation=30)
```
